```python
# 라이브러리 임포트
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# 타이타닉 데이터셋 로드
df = sns.load_dataset('titanic') 

# 인코딩 및 공학 실습에 필요한 컬럼만 선택
df_encode = df[['sex', 'embarked', 'pclass', 'sibsp', 'parch', 'age']]

# 결측치 처리
df_encode['embarked'].fillna(df_encode['embarked'].mode()[0], inplace=True)
df_encode['age'].fillna(df_encode['age'].median(), inplace=True)

df_encode.head()
```

```python
df_encode.info()
```

```python
# 'sex'와 'embarked' 컬럼에 원-핫 인코딩 적용

# Pandas get_dummies() 사용
df_encoded = pd.get_dummies(df_encode, columns=['sex', 'embarked'], drop_first=True, dtype=int)

# drop_first=True: 첫 번째 범주('female' 또는 'S')를 제외하고 N-1개 특성만 생성하여 다중공선성 방지

df_encoded.head()
```
